# Set

```python
prices = [10.0, 11.5, 10.0, 23.5, 11.5, 18.0, 10.0, 10.0]

print('prices list:', prices)
print('prices set:', set(prices))
print('prices set:', set(prices))
```

```python
john_fruits = set(['Apple', 'Cherry', 'Pear'])
jane_fruits = set(['Apple', 'Banana', 'Orange'])
kate_fruits = set(['Apple', 'Banana', 'Cherry', 'Orange', 'Grapes'])
james_fruits = set([])
```

## Difference

```python
john_fruits - jane_fruits
```

```python
kate_fruits - john_fruits - jane_fruits
```

```python
john_fruits - james_fruits
```

## Union

```python
kate_fruits | john_fruits | jane_fruits
```

```python
john_fruits | jane_fruits
```

## Intersection

```python
john_fruits & jane_fruits
```

```python
john_fruits & jane_fruits & kate_fruits
```

```python
john_fruits & jane_fruits & kate_fruits & james_fruits
```

## Symmetric Difference

```python
john_fruits ^ jane_fruits
```

```python
john_fruits ^ kate_fruits
```

```python
print(kate_fruits ^ john_fruits ^ jane_fruits)
print(john_fruits ^ kate_fruits ^ jane_fruits)
print(jane_fruits ^ kate_fruits ^ john_fruits)
print(jane_fruits ^ john_fruits ^ kate_fruits)
```

## Subset & Superset

```python
kate_fruits >= john_fruits
```

```python
kate_fruits >= jane_fruits
```

```python
james_fruits < john_fruits
```
