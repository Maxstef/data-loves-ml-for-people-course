# Uniform Distribution

## Formula

The **probability density function (PDF)** of a continuous uniform distribution on the interval $[a, b]$ is 

$f(x) = \frac{1}{b - a}$, for $a \le x \le b$.


## Characteristics

- All values in the interval \([a, b]\) have the same probability.
- **Mathematical expectation (mean):** $\mu = \frac{a + b}{2}$
- **Variance:** $\sigma^2 = \frac{(b - a)^2}{12}$
- **Interval where the probability density function is constant:** \([a, b]\)

## Examples of Use

- **Random number generation:**  
  When generating a random identification number or password where each character has an equal probability of being any letter or digit.

- **Rolling a die:**  
  The outcome of each roll of a die from 1 to 6 has the same probability.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from scipy.stats import uniform

# Distribution parameters
a = 0
b = 1

# Generate a sample
sample = uniform.rvs(loc=a, scale=b - a, size=1000)
plt.hist(sample)
```

# Exponential Distribution

## Formula

The probability density function (PDF) of the exponential distribution is $f(x; \lambda) = \lambda e^{-\lambda x}, \ x \ge 0$.


## Characteristics

- Used to model the time between events that occur at a constant average rate.
- **Mathematical expectation (mean):** $\mu = \frac{1}{\lambda}$
- **Variance:** $\sigma^2 = \frac{1}{\lambda^2}$
- **Memoryless property:** The probability of an event does not depend on the waiting time.

## Examples of Use

- **Time between events:** The time between website visits by a user or the time between calls to a call center.
- **Time to failure:** The time until a piece of equipment breaks down.


```python
from scipy.stats import expon
import matplotlib.pyplot as plt

# Distribution parameters
scale = 4  # mean = 1/Î»

# Generate a sample
sample = expon.rvs(scale=scale, size=1000)

# Plot histogram
plt.hist(sample, bins=40)
plt.show()
```

```python
log_sample = np.log(sample)
plt.hist(log_sample, bins=40);
```

# Binomial Distribution

## Formula

The **probability mass function (PMF)** of a binomial distribution with parameters $n$ (number of trials) and $p$ (probability of success) is:

$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$, for $k = 0, 1, 2, \dots, n$.

## Characteristics

- Used to model the number of successes in a series of independent trials with two possible outcomes (success/failure).
- **Mathematical expectation (mean):** $\mu = n p$
- **Variance:** $\sigma^2 = n p (1-p)$

## Example Situations

- **Number of successes:** Number of clicks on an advertisement on a website out of the total number of impressions, if the click probability is known.
- **Number of defects:** Number of defective products in a batch.


```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# Distribution parameters
n = 100  # number of trials
p = 0.75  # probability of success

# Generate a sample
sample = binom.rvs(n=n, p=p, size=1000)

# Plot histogram
plt.hist(sample, bins=20)
plt.show()
```

# Poisson Distribution

## Formula

The **probability mass function (PMF)** of the Poisson distribution with parameter $\lambda$ (average number of events) is:

$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}$, for $k = 0, 1, 2, \dots$

## Characteristics

- Used to model the number of events occurring in a fixed interval of time or space.
- **Mathematical expectation (mean):** $\mu = \lambda$
- **Variance:** $\sigma^2 = \lambda$

## Example Situations

- **Number of events:** Number of website visits per hour or number of calls to a call center per day.
- **Rare events:** Number of road accidents per day or number of defects on a production line.


```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Distribution parameter
mu = 5  # average number of events

# Generate a sample
sample = poisson.rvs(mu=mu, size=1000)

# Plot histogram
plt.hist(sample, bins=15)
plt.show()
```
