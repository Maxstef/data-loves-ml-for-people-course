# Logistic regression parctice
This notebook provides practice with logistic regression mathematics and visualizations

## One feature logistic regression

### Almost perfect data

```python
from mlpeople.models.logistic import (
    generate_logistic_regression_data,
    fit_logistic,
    plot_log_loss_curve,
    plot_sigmoid_fit_1d,
)

# Generate data
X, y, true_theta = generate_logistic_regression_data(
    n=100,
    p=1,
    intercept=-0.5,
    coefs=[3],
    seed=42,
    sigma=0.25,
)

# Fit model
theta, history = fit_logistic(X, y, lr=0.2, epochs=300, verbose=False)

# Visualize
plot_log_loss_curve(history)
plot_sigmoid_fit_1d(
    X,
    y,
    theta,
    true_theta=true_theta,
)
```

### A bit noisy data & high learning rate

```python
# More noise 
X, y, true_theta = generate_logistic_regression_data(
    n=100,
    p=1,
    intercept=-0.5,
    coefs=[3],
    seed=42,
    sigma=1,
)

# greater learning rate
theta, history = fit_logistic(X, y, lr=0.5, epochs=300, verbose=False)

# Visualize
plot_log_loss_curve(history)
plot_sigmoid_fit_1d(
    X,
    y,
    theta,
    true_theta=true_theta,
)
```

### Noisy data

```python
# Very noisy data
X, y, true_theta = generate_logistic_regression_data(
    n=100,
    p=1,
    intercept=-0.5,
    coefs=[3],
    seed=42,
    sigma=5,
)

# moderate learning rate
theta, history = fit_logistic(X, y, lr=0.2, epochs=200, verbose=False)

# Visualize
plot_log_loss_curve(history)
plot_sigmoid_fit_1d(
    X,
    y,
    theta,
    true_theta=true_theta,
)
```

## Two feature logistic regression

### Almost perfect data & high learning rate

```python
from mlpeople.models.logistic import (
    generate_logistic_regression_data,
    fit_logistic,
    plot_decision_boundary_2d,
    plot_log_loss_curve,
    plot_predicted_probabilities
)

# Generate data
X, y, _ = generate_logistic_regression_data(p=2, sigma=0.1, seed=42,)

# Fit model
theta, history = fit_logistic(X, y, lr=0.5, epochs=200, verbose=False)

# Visualize
plot_log_loss_curve(history)
plot_decision_boundary_2d(X, y, theta)
plot_predicted_probabilities(X, y, theta)
```

### A bit noisy data & low learning rate

```python
# even more noise
X, y, _ = generate_logistic_regression_data(p=2, sigma=1, seed=77,)

# low learning rate
theta, history = fit_logistic(X, y, lr=0.05, epochs=300, verbose=False)

# Visualize
plot_log_loss_curve(history)
plot_decision_boundary_2d(X, y, theta)
plot_predicted_probabilities(X, y, theta)
```

### Very Noisy data

```python
# even more noise
X, y, _ = generate_logistic_regression_data(p=2, sigma=5, seed=888,)

# hihg learning rate, less epochs
theta, history = fit_logistic(X, y, lr=0.5, epochs=100, verbose=False)

# Visualize
plot_log_loss_curve(history)
plot_decision_boundary_2d(X, y, theta)
plot_predicted_probabilities(X, y, theta)
```
