## Introduction to Recommendation Systems and Motivation

### What is a Recommendation System?

A **recommendation system** suggests products, services, or content that a user is likely to be interested in or interact with.

**Example**

Suppose a user visits Amazon to buy a product. Amazon has information about this user's previous interactions, as well as data from millions of other users with similar preferences and purchasing behavior.

The central question is:

> **What should Amazon recommend to this user to maximize the likelihood of a positive interaction or purchase?**

Unlike many traditional machine learning tasks, there is often **no single correct answer**. There is no explicit label that tells us exactly which item should be recommended.

Instead, the goal is to **learn and understand user behavior** in order to predict which items the user is most likely to find relevant.

---

### Why should we care about Recommendation Systems?

Recommendation systems affect almost every aspect of our digital lives. Whether we realize it or not, they influence much of what we watch, listen to, read, buy, and even who we connect with.

Examples include:

* Music streaming (Spotify)
* Video platforms (YouTube)
* Movies and TV shows (Netflix)
* Online shopping (Amazon, Rozetka, Prom.ua)
* News feeds and social media
* Restaurant recommendations
* Dating applications
* Friend suggestions (Facebook)
* Professional networking (LinkedIn)

Recommendation systems are also a key driver behind the success of many technology companies, including Amazon, Netflix, TikTok, and Meta.

One of their most important roles is to **reduce information overload**.

Modern online platforms often contain millions of items. Without recommendation and ranking algorithms, users would have to manually browse enormous catalogs to find something relevant.

For example, imagine searching for a hair dryer on Rozetka without any personalized recommendations or relevance-based sorting. You might need to browse through thousands of products before finding one that matches your needs.



### What does a Recommendation System actually predict?

A recommendation system does not necessarily try to predict what a user will buy. More generally, it predicts the likelihood of a **user interaction** with an item.

Depending on the application, an interaction may be:

* a click,
* a purchase,
* a movie watch,
* a song play,
* a like or share,
* adding a product to a cart or wishlist.

The exact objective depends on the business. For example, YouTube may optimize for watch time, while Amazon may focus on purchases.

Recommendation systems learn from **historical user behavior**. Instead of asking:

> *"What is the single correct item to recommend?"*

they ask:

> *"Which items is this user most likely to interact with?"*

The goal is to rank available items and present the most relevant ones to the user.


## Limitations of Recommendation Systems

Despite their benefits, recommendation systems also have important limitations.

They often suggest items that are similar to those a user has previously interacted with or that are popular among similar users. While this can improve the user experience, it may also create a **filter bubble**.

A filter bubble reduces a user's exposure to diverse information and viewpoints. This can be particularly problematic in areas such as news consumption, scientific research, and political discussions, where considering multiple perspectives is important.

---

## Data and Main Approaches

### What data do we need to build a Recommendation System?

Recommendation systems can use several types of data:

* **User-item interactions**: purchases, ratings, clicks, likes, adding items to a wishlist, etc.
* **User features**: demographic information or other descriptive attributes.
* **Item features**: category, description, price, genre, and other characteristics.

In the simplest case, a recommendation system can be built using only **historical user-item interactions**, without any additional information about users or items.


## Main Approaches to Building Recommendation Systems

### 1. Collaborative Filtering

Collaborative Filtering is based on a simple idea:

> **Users with similar preferences tend to interact with similar items.**

This approach relies only on **historical user-item interactions** (ratings, clicks, purchases, views) and does not require explicit information about users or items.

The goal is to discover hidden patterns in these interactions and use them to generate recommendations.

There are two main variants of Collaborative Filtering.

---

### User-based Collaborative Filtering

![](https://i0.wp.com/spotintelligence.com/wp-content/uploads/2024/04/user-based-collaborative-filtering.jpg?fit=1200%2C675\&ssl=1)

**Idea:**

Find users with preferences similar to the target user and recommend the items they liked.

#### How it works

1. Compute similarity between users based on their interactions.
2. Find the most similar users.
3. Recommend items they have interacted with but the target user has not.

**Intuition:**

> If people with tastes similar to yours liked an item, you will probably like it too.

---

### Item-based Collaborative Filtering

![](https://i0.wp.com/spotintelligence.com/wp-content/uploads/2024/04/item-based-collaborative-filtering.jpg?resize=1024%2C576\&ssl=1)

**Idea:**

Instead of finding similar users, find items that are similar to those the user already likes.

#### How it works

1. Compute similarity between items based on user interactions.
2. Find items the user has already interacted with.
3. Recommend similar items.

**Intuition:**

> If you liked this product, movie, or course, you will probably like similar ones as well.

---

### What is the main difference?

#### User-based Collaborative Filtering

This approach compares a user with many other users.

As the number of users grows, computing and updating user similarities becomes increasingly expensive. In addition, user preferences often change over time, requiring frequent recalculation.

**Advantages:**

* Simple and intuitive
* Works well on smaller datasets

**Limitations:**

* Difficult to scale
* Computationally expensive for large platforms

---

#### Item-based Collaborative Filtering

This approach focuses on similarities between items instead of users.

The number of items is often smaller than the number of users, and item relationships tend to be more stable over time. As a result, item similarities can be precomputed and reused.

**Advantages:**

* Faster recommendation generation
* Better scalability
* More stable in production systems

---

Both approaches rely solely on **historical interaction data** and do not require explicit user or item features.

**Important**: In Item-based Collaborative Filtering, items are considered similar if they are frequently interacted with by the same users. No explicit item features (such as category or description) are required.

**Limitation:** Collaborative Filtering cannot easily recommend completely new users or completely new items because there is little or no interaction history available. This challenge is known as the cold start problem.


### 2. Content-Based Recommender Systems

Content-Based Recommendation Systems focus on the **properties of the items themselves** and the preferences of a particular user.

The main idea is simple:

> **Recommend items that are similar to those the user already liked.**

![](https://i0.wp.com/spotintelligence.com/wp-content/uploads/2023/11/content-based-recommendation-system-1024x576.webp?resize=1024%2C576\&ssl=1)

---

### How does it work?

The first step is to represent each item using its features.

Examples:

* **Movies:** genre, actors, director, description.
* **Products:** category, brand, price, technical specifications.
* **Articles:** keywords, topics, text content.

These features are then converted into a numerical representation that a machine learning model can process.

Depending on the type of content, different techniques can be used:

* **Text:** TF-IDF or text embeddings.
* **Images:** features extracted by computer vision models.
* **Structured data:** numerical or categorical features.

In modern recommendation systems, these representations are often learned automatically as **embeddings**.

---

### How are recommendations generated?

The system:

1. Builds a user profile from items the user has previously interacted with.
2. Finds items with similar content.
3. Recommends those items to the user.

Unlike Collaborative Filtering, this approach does not depend on the behavior of other users.

---

### Advantages and Limitations

**Advantages:**

* Handles the **cold start problem for new items**.
* Produces personalized recommendations.
* Does not require data from other users.

**Limitations:**

* May recommend only very similar content, reducing diversity.
* Cannot easily discover unexpected but relevant items based on community behavior.


| Collaborative Filtering               | Content-Based Filtering            |
| ------------------------------------- | ---------------------------------- |
| Learns from **user behavior**         | Learns from **item features**      |
| No item descriptions required         | Requires item features             |
| Can discover surprising relationships | Usually recommends similar content |
| Suffers from new-item cold start      | Handles new-item cold start well   |


### 3. Hybrid Recommendation Systems

Collaborative Filtering and Content-Based Filtering each have their own strengths and limitations.

* **Content-Based Filtering** works well for new items but may recommend only similar content.
* **Collaborative Filtering** can discover patterns from community behavior but requires sufficient interaction history.

For this reason, most modern recommendation systems use a **hybrid approach** that combines both methods.

A hybrid model can leverage:

* **user behavior** (Collaborative Filtering),
* **item features** (Content-Based Filtering).

This combination helps to:

* reduce the **cold start problem**,
* improve recommendation quality,
* produce more robust results across different scenarios.

---

### How do Hybrid Systems work?

Hybrid recommendations can be generated in several ways:

* combining the outputs of multiple models;
* using content features together with interaction data;
* enriching collaborative recommendations with content-based signals.

As a result, the system benefits from both **individual user preferences** and **collective behavior patterns**.

---

Today, most large-scale platforms use some form of **hybrid recommendation system**, combining multiple data sources and machine learning models to deliver better recommendations.


## Formulating the Recommendation Problem

Recommendation systems are typically built from **user-item interactions**.

We have two entities:

* **(N) users**
* **(M) items**

Users are the consumers, while items are the products or services being recommended.

Examples of items:

* Movies (Netflix)
* Books (Amazon)
* Songs (Spotify)
* Products (Rozetka)
* User profiles (Tinder)

---

### The Utility Matrix

The interactions between users and items are represented by a **Utility Matrix**.

Each element ($y_{ij}$) represents the interaction between user ($i$) and item ($j$).

Depending on the application, an interaction may correspond to:

* a rating,
* a click,
* a purchase,
* a view,
* or any other user action.

For example, the matrix below contains (N=5) users and (M=6) items.

![](https://takuti.github.io/Recommendation.jl/latest/assets/images/cf.png)

Rows represent users, columns represent items.

A recommendation system can therefore represent:

* users by the items they interact with,
* items by the users who interact with them.

---

### Why is the Utility Matrix Sparse?

In real-world applications, users interact with only a small fraction of all available items.

For example, an online store may offer thousands of products, but a typical customer purchases only a few of them.

As a result, most entries in the utility matrix are missing, making it a **highly sparse matrix**.

---

### What are we trying to predict?

The central goal of a recommendation system is to **estimate the missing values in the Utility Matrix**.

In other words, given the observed interactions, we want to predict which unseen items a user is likely to prefer.

This problem is often referred to as **matrix completion**.

Once the missing interactions or ratings are estimated, we can recommend the items with the highest predicted scores.

---

### How is this different from traditional Machine Learning?

In many supervised learning problems:

* we have a feature matrix (X),
* and a target vector (y).

The objective is to predict the target variable.

In recommendation systems, the data is organized differently. Instead of predicting a single target column, we aim to predict the **missing entries of the Utility Matrix (Y)**.

This unique structure is one of the main characteristics of recommendation systems.


**Key Idea**: Recommendation systems are often viewed as a matrix completion problem—we observe only a small fraction of user-item interactions and try to infer the missing ones.
