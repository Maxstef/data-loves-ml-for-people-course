# Central Limit Theorem Experiments

## Proportions

### Flip Coin Experiment
Flip a fair coin 100 times and record the number of heads. Repeat this experiment 1,000 times. Then, plot the distribution of the proportion of heads obtained in each experiment.

```python
clt_sample_proportions(bins=20, random_seed=1);
```

### Dice rolling experiment
Roll a fair die 100 times and record the number of sixes. Repeat this experiment 1,000 times, then plot the distribution of the proportion of sixes obtained in each experiment.

```python
clt_sample_proportions(bins=20, outcome_count=6, random_seed=1);
```

## Means

### Dice rolling experiment for means
Roll a fair die N times and compute the average (mean) value of the outcomes. Repeat this experiment 1,000 times and plot the distribution of the sample means.

Repeat the entire procedure starting with N=1, then increase N to 2, 3, 5, and so on. As N increases, the distribution of the sample mean becomes increasingly close to a normal distribution. As the sample size grows, the distribution of the mean approaches the population mean μ, and its variance approaches 𝜎^2/𝑁, where N is the sample size.

```python
clt_sample_means(sample_size=1, bins=6, random_seed=48)
clt_sample_means(sample_size=2, bins=11, random_seed=48)
clt_sample_means(sample_size=3, bins=14, random_seed=48)
clt_sample_means(sample_size=5, bins=18, random_seed=48)
clt_sample_means(sample_size=10, random_seed=48)
clt_sample_means(sample_size=20, random_seed=48)
clt_sample_means(sample_size=40, random_seed=48)
clt_sample_means(sample_size=160, random_seed=48);
```

## Back to Proportions

### Sample Proportions from a Categorical Population (Biased dice)
Roll a dice that has more probability of getting large number.

Generate a categorical population with specified outcome probabilities. Repeatedly draw samples of fixed size and compute the proportion of a target category in each sample. Repeat this experiment 1,000 times and plot the distribution of sample proportions. As the sample size increases, the distribution of the sample proportion approaches a normal distribution with mean p and variance p(1−p)/n.

```python
# Biased dice
values = [1, 2, 3, 4, 5, 6]
probabilities = [0.05, 0.1, 0.15, 0.2, 0.25, 0.25]

clt_from_categorical_population(
    values=values,
    probabilities=probabilities,
    target_value=6,
    sample_size=60,
    bins=20,
    random_seed=42
);
```
