1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 10 * x - 5

def derivative(f, x, h=1e-6):
    return (f(x + h) - f(x)) / h

def plot_f_and_derivative(
    f, xs, 
    title='Graph of a function and its derivative',
    xlabel='x', ylabel='y',
    flabel=None, fdlabel=None
):
    flabel = 'f(x) = ' + flabel if flabel is not None else 'f(x)'
    fdlabel = "f'(x) = " + fdlabel if fdlabel is not None else "f'(x)"

    plt.figure(figsize=(10, 5))
    plt.plot(xs, f(xs), label=flabel, color='blue')
    plt.plot(xs, derivative(f, xs), label=fdlabel, color='red', linestyle='--')

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)

    plt.show()


xs = np.linspace(0, 10, 100)
# np.gradient(f(xs), xs)

plot_f_and_derivative(f, xs, xlabel='learn hours', ylabel='final score', flabel='10x+5', fdlabel='10')

```

<span style="background-color: #f60">Відповідь:</span><br/>
 - маємо лінійну залежність - зі збільшення кількості годин збільшується оцінка
 - оскільки за кожну додаткову годину оцінка збільшується на 10 балів, швидкість зміни є сталою. Тому похідна цієї функції дорівнює 10 у всіх точках.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp

def f_sin_fn(x):
    return np.sin(x)

def f_ln_fn(x):
    return np.log(x)

x = sp.symbols('x')

f_sin = sp.sin(x)
f_ln = sp.log(x)

df_sin = sp.diff(f_sin, x)
df_ln = sp.diff(f_ln, x)

print(f'function: {f_sin}, derivative: {df_sin}')
print(f'function: {f_ln}, derivative: {df_ln}')

x_sin_vals = np.linspace(-2*np.pi, 2*np.pi, 400)
x_ln_vals = np.linspace(0.01, 5, 400)

plot_f_and_derivative(f_sin_fn, x_sin_vals, flabel=str(f_sin), fdlabel=str(df_sin))
plot_f_and_derivative(f_ln_fn, x_ln_vals, flabel=str(f_ln), fdlabel=str(df_ln))
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

f_x = x**2 - 5*x + 1
derivative_f_x = sp.diff(f_x, x)

print(f'function: {f_x}, derivative: {derivative_f_x}')

xs = np.linspace(0, 100, 100)

plot_f_and_derivative(sp.lambdify(x, f_x, 'numpy'), xs, flabel=str(f_x), fdlabel=str(derivative_f_x))
```
