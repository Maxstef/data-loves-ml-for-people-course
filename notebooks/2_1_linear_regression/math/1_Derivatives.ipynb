```python
import sympy as sp

def f_sin_fn(x):
    return np.sin(x)

def f_ln_fn(x):
    return np.log(x)

x = sp.symbols('x')

f_sin = sp.sin(x)
f_ln = sp.log(x)

df_sin = sp.diff(f_sin, x)
df_ln = sp.diff(f_ln, x)

print(f'function: {f_sin}, derivative: {df_sin}')
print(f'function: {f_ln}, derivative: {df_ln}')

x_sin_vals = np.linspace(-2*np.pi, 2*np.pi, 400)
x_ln_vals = np.linspace(0.1, 10, 400)

plot_function_and_numeric_derivative(
    f_sin_fn,
    x_sin_vals,
    function_label=str(f_sin),
    derivative_label=str(df_sin)
)

plot_function_and_numeric_derivative(
    f_ln_fn,
    x_ln_vals,
    function_label=str(f_ln),
    derivative_label=str(df_ln)
);
```
