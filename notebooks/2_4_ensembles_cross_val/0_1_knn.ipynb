# K Nearest Neighbors (KNN)

In this notebook there is implementation of KNN from scratch



```python
from mlpeople.models.knn.knn import KNN
from mlpeople.models.knn.data import generate_train_test_data
from mlpeople.models.knn.visualization import visualize_knn_classifier
```

## Generate synthetic 2D dataset

```python
X_train, X_test, y_train, y_test = generate_train_test_data(n_samples=33)
```

## Visualize neighbors of specific test point

```python
visualize_knn_classifier(
    X_train, X_test, y_train, y_test,
    k=5, p=2,
    show_neighbors=True,
    test_point=X_test[0]
)
```

```python
visualize_knn_classifier(
    X_train, X_test, y_train, y_test,
    k=5, p=2,
    show_neighbors=True,
    test_point=X_test[1]
)
```

```python
visualize_knn_classifier(
    X_train, X_test, y_train, y_test,
    k=5, p=2,
    show_neighbors=True,
    test_point=X_test[2]
)
```

## Euclidean Distance (p = 2)

```python
for k in range(2, 7):
    visualize_knn_classifier(X_train, X_test, y_train, y_test, k=k, p=2)
```

## Manhattan Distance (p = 1)

```python
for k in range(2, 7):
    visualize_knn_classifier(X_train, X_test, y_train, y_test, k=k, p=1)
```

## Minkowski p=5

```python
for k in range(4, 8):
    visualize_knn_classifier(X_train, X_test, y_train, y_test, k=k, p=5)
```
