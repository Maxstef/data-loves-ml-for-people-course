# PCA Explorer Implementation

This notebook is a practice and test of customly implemented PCAExplorer class

## Wine Dataset

```python
%load_ext autoreload
%autoreload 2

from sklearn.datasets import load_wine
import numpy as np

from mlpeople.statistic.pca_explorer import PCAExplorer

wine = load_wine()
X = wine.data
y = wine.target

mean_wine = np.mean(X, axis=0)
std_wine = np.std(X, axis=0)

X_wine_std = (X - mean_wine) / std_wine

explorer = PCAExplorer(n_components=3)

explorer.fit(X_wine_std)

explorer.plot_explained_variance()
explorer.plot_cumulative_variance()
explorer.plot_variance_bars()

explorer.plot_3d(y)

explorer.plot_loadings(wine.feature_names, pc=0)
explorer.plot_loadings(wine.feature_names, pc=1)

explorer.plot_reconstruction_error(X, max_components=10)
```

## Digits Dataset

```python
from sklearn.datasets import load_digits

digits = load_digits()
X_digits = digits.data
y_digits = digits.target

explorer_d = PCAExplorer(n_components=3)

explorer_d.fit(X_digits)

explorer_d.plot_explained_variance()
explorer_d.plot_cumulative_variance()
explorer_d.plot_variance_bars()

explorer_d.plot_3d(y_digits)

explorer_d.plot_reconstruction_error(X_digits, max_components=20)
```
