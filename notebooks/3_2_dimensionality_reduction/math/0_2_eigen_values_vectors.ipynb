## Step 1 - What is the problem?

We start with a square matrix $ A $ and ask:

> Does there exist a non-zero vector whose direction is unchanged by $ A $?

Mathematically:

$
A v = \lambda v
$

We are looking for:
- $ v \neq 0 $
- direction is preserved
- only scaling changes

---

## Step 2 - Geometric meaning

A [matrix transformation](https://en.wikipedia.org/wiki/Transformation_matrix) usually:
- rotates vectors
- stretches them differently in different directions
- distorts geometry

But eigenvectors are **special directions** where the transformation behaves like pure scaling:

$
A v \parallel v
$

That means:
> after applying $A$, the vector still lies on the same line through the origin

````python
import numpy as np

A = np.array([
    [0, 5, -10],
    [0, 22, 16],
    [0, -9, -2],
])

X1 = np.array([-5, -4, 3])
X2 = np.array([1, 0, 0])
````

````python
A @ X1, X1 * 10
````

````python
A @ X2, X2 * 0
````

## Matrix–Vector Multiplication Example

We are given:

$$
A =
\begin{bmatrix}
0 & 5 & -10 \\
0 & 22 & 16 \\
0 & -9 & -2
\end{bmatrix}
\quad
X_1 =
\begin{bmatrix}
-5 \\
-4 \\
3
\end{bmatrix}
$$

We want to compute:

$$
A X_1
$$

---

**Step 1 — [Rule of matrix multiplication](https://en.wikipedia.org/wiki/Matrix_multiplication#Matrix_times_vector)**

Each entry of the result is computed as a **[dot product](https://en.wikipedia.org/wiki/Dot_product) between a row of $A$** and the vector $X_1$.

---

**Step 2 — First row**

$$
[0,\ 5,\ -10] \cdot [-5,\ -4,\ 3]
$$

$$
= 0 \cdot (-5) + 5 \cdot (-4) + (-10) \cdot 3
$$

$$
= 0 - 20 - 30 = -50
$$

---

**Step 3 — Second row**

$$
[0,\ 22,\ 16] \cdot [-5,\ -4,\ 3]
$$

$$
= 0 \cdot (-5) + 22 \cdot (-4) + 16 \cdot 3
$$

$$
= 0 - 88 + 48 = -40
$$

---

**Step 4 — Third row**

$$
[0,\ -9,\ -2] \cdot [-5,\ -4,\ 3]
$$

$$
= 0 \cdot (-5) + (-9) \cdot (-4) + (-2) \cdot 3
$$

$$
= 0 + 36 - 6 = 30
$$

---

**Final result**

$$
A X_1 =
\begin{bmatrix}
-50 \\
-40 \\
30
\end{bmatrix}
$$

---

**Key insight (preview of eigenvectors)**

Notice that:

$$
A X_1 = 10 \cdot X_1
$$

This means:
- the vector direction does not change
- only its length is scaled by 10

So $X_1$ behaves like an **eigenvector of $A$** with eigenvalue $ \lambda = 10 $.

## Step 3 - Algebraic Form

We start from the eigenvalue equation:

$
A v = \lambda v
$

The goal is to find vectors $v \neq 0$ whose direction does not change under the transformation by $A$.

To make this easier to analyze, move everything to one side:

$
A v - \lambda v = 0
$

Factor out $v$:

$
(A - \lambda I) v = 0
$

where $I$ is the [identity matrix](https://en.wikipedia.org/wiki/Identity_matrix).

This transforms the eigenvalue problem into a [**homogeneous linear system**](https://www.bohrium.com/en/sciencepedia/feynman/linear_algebra_undergraduate-homogeneous_linear_systems).

---

## Step 4 - When does a non-trivial solution exist?

The equation

$
(A - \lambda I) v = 0
$

always has the [**trivial solution**](https://www.cuemath.com/algebra/homogeneous-system-of-linear-equations/):

$
v = 0
$

But we are not interested in this.  
We want a **non-zero vector** $v \neq 0$.

From linear algebra, a homogeneous system has [non-trivial solutions](https://amixsystems.com/homogeneous-system/#solution-types). **only when the matrix is not [invertible](https://en.wikipedia.org/wiki/Invertible_matrix)**

[A matrix is not invertible when its determinant is zero](https://math.stackexchange.com/questions/2093067/why-do-homogenous-linear-systems-have-determinant-zero):

$
\det(A - \lambda I) = 0
$

This equation is called the [**characteristic equation**](https://en.wikipedia.org/wiki/Characteristic_polynomial).

Solving it gives the values of $ \lambda $ for which eigenvectors exist.

These values are called **eigenvalues**.

## Step 5 - Eigenvalues Come First (Characteristic Equation)

From the previous step, we know that a non-trivial solution to

$
(A - \lambda I)v = 0
$

exists only when

$
\det(A - \lambda I) = 0
$

This determinant produces a **polynomial in $ \lambda $** called the:

> [**characteristic polynomial**](https://en.wikipedia.org/wiki/Characteristic_polynomial)

Solving this equation gives the values of $ \lambda $ for which eigenvectors exist.

These values are called **eigenvalues** of the matrix $A$.

So the process always starts with:

$
\det(A - \lambda I) = 0
\quad \Longrightarrow \quad
\text{find eigenvalues}
$

---


## Step 6 - Eigenvectors Come Next

Once an eigenvalue $ \lambda $ is known, we substitute it back into:

$
(A - \lambda I)v = 0
$

This becomes a standard homogeneous linear system.

Solving this system gives the corresponding **eigenvectors**.

So the overall procedure is:

1. Compute $ \det(A - \lambda I) $
2. Solve for eigenvalues $ \lambda $
3. For each $ \lambda $, solve $ (A - \lambda I)v = 0 $ to find eigenvectors

## Step 7 - Key Interpretation

An eigenpair $ (\lambda, v) $ satisfies:

$
A v = \lambda v
$

This means:

- The vector $v$ keeps its **direction** after the transformation by $A$
- The scalar $ \lambda $ tells how much the vector is **stretched or shrunk**

So, along eigenvectors, the matrix transformation behaves like **pure scaling**.

In other words:

> A matrix can be understood by the special directions where it acts as simple multiplication.

## Finding Eigenvalues and Eigenvectors - Worked Example

Consider the matrix

$
A =
\begin{bmatrix}
-5 & 2 \\
-7 & 4
\end{bmatrix}
$

We will now follow the exact procedure developed earlier:

1. Form $ A - \lambda I $
2. Compute $ \det(A - \lambda I) $ — the characteristic equation
3. Solve for eigenvalues $ \lambda $
4. Substitute each eigenvalue into $ (A - \lambda I)v = 0 $ to find eigenvectors

````python
# Step 1 — Define the matrix
A = np.array([
    [-5, 2],
    [-7, 4],
])
````

````python
# Step 2 — Compute A−λI symbolically
import sympy as sp

lam = sp.symbols('lam')

A_sym = sp.Matrix([
    [-5, 2],
    [-7, 4]
])

I = sp.eye(2)

A_minus_lamI = A_sym - lam * I
A_minus_lamI
````

````python
# Step 3 — Characteristic equation det(A−λI)=0
char_poly = A_minus_lamI.det()
char_poly
````

````python
# Step 4 — Solve for eigenvalues λ^2+λ-6=0
eigenvalues = sp.solve(char_poly, lam)
eigenvalues
````

````python
# Step 5 — Compute eigenvectors as (λI - A) X = 0 where X is eigenvector
# (-3I - A) X = 0 - first 
# (2I - A) X = 0 - second 
eigenvectors = A_sym.eigenvects()
eigenvectors
````

**Clean summary**
| eigenvalue | multiplicity | eigenvector |
| ---------- | ------------ | ----------- |
| -3         | 1            | ([1, 1])    |
| 2          | 1            | ([2/7, 1])  |

second vector could be multiplied by 7 and get [2, 7] - still a perfectly valid eigenvector

````python
# Step 6 — Verify using NumPy (numeric check)
eigvals_np, eigvecs_np = np.linalg.eig(A)
eigvals_np, eigvecs_np
````

[-0.70710678, -0.70710678] the same as [1, 1]

[-0.27472113, -0.96152395] the same as [2/7, 1]

-0.2747 / -0.9615 ≈ 0.2857 ≈ 2 / 7

````python
for i in range(len(eigvals_np)):
    v = eigvecs_np[:, i]
    lam = eigvals_np[i]

    lhs = A @ v
    rhs = lam * v

    print(f"\nEigenvalue {lam}")
    print("A v =", lhs)
    print("λ v =", rhs)
    print("close?", np.allclose(lhs, rhs))
````

### Important Note - Eigenvector Scaling

Eigenvectors are **not unique vectors**, they are **directions**.

If $v$ is an eigenvector of $A$, then any scaled version is also an eigenvector:

$
A v = \lambda v \quad \Rightarrow \quad A (c v) = \lambda (c v)
$

for any non-zero scalar $c$.

---

**Example**

These vectors represent the **same eigenvector direction**:

$
\begin{bmatrix} 1 \\ 1 \end{bmatrix}
\quad \text{and} \quad
\begin{bmatrix} -0.707 \\ -0.707 \end{bmatrix}
$

They differ only by scaling, but define the same geometric object:

> a line (1D subspace) through the origin.

---

**What changes and what does not**

| Representation | Meaning |
|------|--------|
| \([1, 1]\) | Eigenvector direction (not normalized) |
| \([-0.707, -0.707]\) | Same direction, unit-length version |
| both | Same geometric line (same eigenspace) |

---

**What PCA uses**

PCA uses **unit eigenvectors** (orthonormal basis):

- ensures clean projections
- makes variance interpretation valid
- removes scaling ambiguity

So in PCA:

> only direction matters, not magnitude

## A zero eigenvalue?

Given:

$
A =
\begin{bmatrix}
2 & 2 & -2 \\
1 & 3 & -1 \\
-1 & 1 & 1
\end{bmatrix}
$

---

**1. Key idea**

A matrix has a **zero eigenvalue** if:

$
\det(A) = 0
$

because this implies that $ \lambda = 0 $ is a root of:

$
\det(A - \lambda I) = 0
$

---

**2. How to check**

We can check this in two equivalent ways:

**(a) [Determinant](https://en.wikipedia.org/wiki/Determinant) test**

If:

$
\det(A) = 0
$

then the matrix has a zero eigenvalue.

**(b) Eigenvalue computation**

Compute:

$
\text{eigvals}(A)
$

and check whether 0 (or a very small value close to 0) appears.

---

**3. Geometric interpretation**

If $ \lambda = 0 $ is an eigenvalue, then there exists a non-zero vector $v$ such that:

$
A v = 0
$

This means:

- the matrix **collapses a direction to zero**
- the transformation is **not invertible**
- the matrix loses one effective dimension

---

**4. Impact on PCA**

In PCA (applied to a covariance matrix):

- eigenvalues represent **variance along principal directions**
- eigenvectors represent **principal directions**

If $ \lambda = 0 $:

- there is **no variance in that direction**
- the data lies in a **lower-dimensional subspace**
- that principal component carries **no information**

So in PCA:

> a zero eigenvalue means the corresponding principal component can be completely discarded without losing information

---

**5. Summary rule**

| Condition | Meaning |
|------|--------|
| $ \det(A) = 0 $ | zero eigenvalue exists |
| $ \det(A) \neq 0 $ | no zero eigenvalue |
| <a href="https://en.wikipedia.org/wiki/Rank_(linear_algebra)">rank</a> < n | zero eigenvalue guaranteed |

---

**6. Intuition**

A zero eigenvalue means:

> there is a direction in space that gets completely flattened to the origin by the transformation

````python
A = np.array([
    [2, 2, -2],
    [1, 3, -1],
    [-1, 1, 1],
])

np.linalg.eigvals(A).round(0)
````
