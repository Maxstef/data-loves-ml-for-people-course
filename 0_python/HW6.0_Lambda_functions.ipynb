```python
test_y = [1, 1, 0, 0, 1, 1, 0, 0]
pred_y = [1, 1, 0, 1, 1, 0, 0, 0]
```

```python
zip(pred_y, test_y)
```

```python
list(zip(pred_y, test_y))
```

```python
def evaluate_model(predictions, actuals):
    correct_count = sum([1 for p, a in zip(predictions, actuals) if p == a])
    total_count = len(actuals)
    return correct_count / total_count

evaluate_model(pred_y, test_y)
```

```python
evaluate_model(['a', 'b', 'c', 'd', 'd'], ['a', 'b', 'c', 'd', 'e'])
```

```python
import random

possible_values = ['a', 'b', 'c']
actual_values = []
predicted_values = []

for n in range(1000):
    actual_values.append(possible_values[random.randint(0, 2)])
    predicted_values.append(possible_values[random.randint(0, 2)])

evaluate_model(predicted_values, actual_values)
```

```python
student_grades = [
    {
        'name': 'John',
        'subject': 'Math',
        'grade': 60
    },
    {
        'name': 'Hanna',
        'subject': 'Chemistry',
        'grade': 75
    },
    {
        'name': 'Kate',
        'subject': 'Math',
        'grade': 70
    },
    {
        'name': 'John',
        'subject': 'Chemistry',
        'grade': 65
    },
    {
        'name': 'Hanna',
        'subject': 'Math',
        'grade': 92
    },
    {
        'name': 'Kate',
        'subject': 'Chemistry',
        'grade': 90
    }
]
```

```python
def get_sorted_by(items=student_grades, attr='grade', reverse=False):
    return sorted(items, key=lambda item: item[attr], reverse=reverse)
```

```python
get_sorted_by()
```

```python
get_sorted_by(reverse=True)
```

```python
get_sorted_by(attr='name')
```

```python
get_sorted_by(attr='subject')
```

```python
def get_sorted_by_multiple(items=student_grades, attrs=('subject', 'grade'), reverse=False):
    return sorted(items, key=lambda item: tuple(item[attr] for attr in attrs), reverse=reverse)
```

```python
get_sorted_by_multiple(reverse=True)
```

```python
def get_student(name):
    return list(filter(lambda item: item['name'] == name, student_grades))

def get_grades_above_included(n):
    return list(filter(lambda item: item['grade'] >= n, student_grades))

def get_grades_below(n):
    return list(filter(lambda item: item['grade'] < n, student_grades))

def get_changed(attr, from_value, to_value):
    return list(map(lambda item: {**item, attr: to_value} if item[attr] == from_value else item, student_grades))
```

```python
get_student('John')
```

```python
get_grades_above_included(90)
```

```python
get_grades_below(75)
```

```python
get_changed('name', from_value='John', to_value='Dave')
```
