**1. Цикл while**

Задано змінну `counter`, яка рівна -3.

Напишіть програму, використовуючи `while`, яка в циклі додає до змінної `counter` +1 і зупиняється, коли змінна `counter` стане суворо більше за 5. На кожній ітерації циклу виводьте значення змінної `counter`.



```python
counter = -3

while counter <= 5:
    print(counter)
    counter += 1
```

**2. List comprehension - прості операції**

Дано список чисел:

```python
numbers = [11, 4, 9, 16, 7, 5]
```


Створіть нові списки використовуючи list comprehension, які містять (кожен пункт оцінюється в 0.5 бала):
- Квадрати всіх чисел
- Тільки парні числа
- Всі числа помножені на 10

Виведіть новоутворені списки.

```python
square_numbers = [n*n for n in numbers]
even_numbers = [n for n in numbers if n%2 == 0]
mult_by_ten_numbers = [n*10 for n in numbers]

print(square_numbers)
print(even_numbers)
print(mult_by_ten_numbers)
```

**3. List comprehension - комбінована логіка**

Дано список чисел від 1 до 10. Створіть новий список, де виконується наступна логіка:
- Якщо число парне - помножте його на 2
- Якщо непарне - віднімніть 1

Виведіть результуючий список на екран.

```python
numbers = list(range(1, 11))
```

```python
result_numbers = [n*2 if n%2 == 0 else n-1 for n in numbers]

print(result_numbers)
```
